```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
import patsy
```

#### Modelo 1

```python
reg = smf.ols('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda) + C(educacao) + C(estado_civil) + idade + tempo_emprego + qt_pessoas_residencia', data = df).fit()
reg.summary()
```

#### Modelo 2

```python
reg = smf.ols('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(tipo_renda, Treatment(1)) + C(estado_civil) + idade + tempo_emprego + qt_pessoas_residencia', data = df).fit()
reg.summary()
```

#### Modelo 3

```python
reg = smf.ols('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + C(estado_civil, Treatment(2) ) + idade + tempo_emprego + qt_pessoas_residencia', data = df).fit()
reg.summary()
```

#### Modelo 4

```python
reg = smf.ols('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + idade + tempo_emprego + qt_pessoas_residencia', data = df).fit()
reg.summary()
```

#### Modelo 5

```python
reg = smf.ols('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + qtd_filhos + idade + tempo_emprego', data = df).fit()
reg.summary()
```
